## Problem 1: Modular Arithmetic
Find $x$ for the following congruences.

### Part A: $x \equiv 457 \pmod{13}$

We divide 457 by 13 and find the remainder:

$$457 = 35 \times 13 + 2$$

Since $35 \times 13 = 455$ and $457 - 455 = 2$:

$$\boxed{x = 2}$$

### Part B: $x \equiv -25 \pmod{7}$

We first find $25 \pmod{7}$:

$$25 = 3 \times 7 + 4 \implies 25 \equiv 4 \pmod{7}$$

Then negate it and adjust to be positive:

$$-25 \equiv -4 \equiv 7 - 4 \equiv 3 \pmod{7}$$

$$\boxed{x = 3}$$

```python
# Verification - Problem 1
print("Problem 1 Verification:")
print(f"457 mod 13 = {457 % 13}")
print(f"-25 mod 7  = {-25 % 7}")
```

## Problem 2: The Euclidean Algorithm
Calculate $\gcd(1160, 480)$ and find $x, y$ such that $1160x + 480y = \gcd(1160, 480)$.

### Part A: Euclidean Algorithm — $\gcd(1160, 480)$

We apply the algorithm step by step:

| Step | Equation                        | Remainder |
|------|---------------------------------|-----------|
| 1    | $1160 = 2 \times 480 + 200$    | 200       |
| 2    | $480  = 2 \times 200 + 80$     | 80        |
| 3    | $200  = 2 \times 80  + 40$     | 40        |
| 4    | $80   = 2 \times 40  + 0$      | 0         |

The last non-zero remainder is **40**, therefore:

$$\boxed{\gcd(1160, 480) = 40}$$

### Part B: Extended Euclidean Algorithm

We back-substitute to express $\gcd = 40$ as a linear combination of 1160 and 480.

**Step 1:** From row 3:
$$40 = 200 - 2 \times 80$$

**Step 2:** Substitute $80 = 480 - 2 \times 200$ (from row 2):
$$40 = 200 - 2 \times (480 - 2 \times 200) = 5 \times 200 - 2 \times 480$$

**Step 3:** Substitute $200 = 1160 - 2 \times 480$ (from row 1):
$$40 = 5 \times (1160 - 2 \times 480) - 2 \times 480 = 5 \times 1160 - 12 \times 480$$

$$\boxed{x = 5, \quad y = -12}$$

**Verification:** $1160 \times 5 + 480 \times (-12) = 5800 - 5760 = 40$ ✓

```python
# Verification - Problem 2
import math

print("Problem 2 Verification:")
print(f"gcd(1160, 480) = {math.gcd(1160, 480)}")

x, y = 5, -12
print(f"1160*{x} + 480*({y}) = {1160*x + 480*y}")
```

## Problem 3: Euler's Totient Function

Recall: For $n = p_1^{a_1} \cdot p_2^{a_2} \cdots p_k^{a_k}$:
$$\phi(n) = n \cdot \prod_{p \mid n} \left(1 - \frac{1}{p}\right)$$

### Part A: $\phi(21)$

Prime factorization: $21 = 3 \times 7$

$$\phi(21) = 21 \times \left(1 - \frac{1}{3}\right) \times \left(1 - \frac{1}{7}\right) = 21 \times \frac{2}{3} \times \frac{6}{7} = \boxed{12}$$

### Part B: $\phi(17)$

17 is **prime**, so:

$$\phi(17) = 17 - 1 = \boxed{16}$$

### Part C: $\phi(360)$

Prime factorization: $360 = 2^3 \times 3^2 \times 5^1$

$$\phi(360) = 360 \times \left(1 - \frac{1}{2}\right) \times \left(1 - \frac{1}{3}\right) \times \left(1 - \frac{1}{5}\right)$$

$$= 360 \times \frac{1}{2} \times \frac{2}{3} \times \frac{4}{5} = \boxed{96}$$

```python
# Verification - Problem 3
from sympy import totient

print("Problem 3 Verification:")
print(f"phi(21)  = {totient(21)}")
print(f"phi(17)  = {totient(17)}")
print(f"phi(360) = {totient(360)}")
```

## Problem 4: Information Theory — Shannon Entropy

A source produces $\{A, B, C, D\}$ with probabilities:
$$P(A) = \frac{1}{2}, \quad P(B) = \frac{1}{4}, \quad P(C) = \frac{1}{8}, \quad P(D) = \frac{1}{8}$$

The Shannon Entropy is defined as:
$$H(X) = -\sum_{i} P(x_i) \log_2 P(x_i)$$

Computing each term:

| Symbol | $P(x)$ | $\log_2 P(x)$ | $-P(x) \log_2 P(x)$ |
|--------|--------|----------------|----------------------|
| A      | $1/2$  | $-1$           | $1/2$                |
| B      | $1/4$  | $-2$           | $1/2$                |
| C      | $1/8$  | $-3$           | $3/8$                |
| D      | $1/8$  | $-3$           | $3/8$                |

$$H(X) = \frac{1}{2} + \frac{1}{2} + \frac{3}{8} + \frac{3}{8} = \frac{4}{8} + \frac{4}{8} + \frac{3}{8} + \frac{3}{8} = \frac{14}{8} = \boxed{1.75 \text{ bits}}$$

```python
# Verification - Problem 4
import math

probs = [1/2, 1/4, 1/8, 1/8]
H = -sum(p * math.log2(p) for p in probs)

print("Problem 4 Verification:")
print(f"H(X) = {H} bits")
```

## Problem 5: Bayes' Theorem — IDS Attack Detection

**Given:**
- $P(A) = 0.001$ — probability of an actual attack
- $P(F \mid A) = 0.99$ — True Positive rate
- $P(F \mid \neg A) = 0.01$ — False Positive rate

**Find:** $P(A \mid F)$ — probability that an attack is actually occurring given the IDS flagged it.

**Step 1:** Apply the Total Probability Theorem to find $P(F)$:
$$P(F) = P(F \mid A) \cdot P(A) + P(F \mid \neg A) \cdot P(\neg A)$$
$$P(F) = 0.99 \times 0.001 + 0.01 \times 0.999 = 0.00099 + 0.00999 = 0.01098$$

**Step 2:** Apply Bayes' Theorem:
$$P(A \mid F) = \frac{P(F \mid A) \cdot P(A)}{P(F)} = \frac{0.99 \times 0.001}{0.01098} = \frac{0.00099}{0.01098}$$

$$\boxed{P(A \mid F) \approx 0.0902 \approx 9.02\%}$$

**Interpretation:** Even if the IDS flags an event, there is only a ~9% chance that it is an actual attack. This is known as the **base rate fallacy** — when attacks are very rare, even a highly accurate IDS will produce many false positives.

```python
# Verification - Problem 5
P_A = 0.001
P_F_given_A = 0.99
P_F_given_notA = 0.01
P_notA = 1 - P_A

P_F = P_F_given_A * P_A + P_F_given_notA * P_notA
P_A_given_F = (P_F_given_A * P_A) / P_F

print("Problem 5 Verification:")
print(f"P(F)       = {P_F:.5f}")
print(f"P(A | F)   = {P_A_given_F:.4f}")
print(f"As percent = {P_A_given_F * 100:.2f}%")
```
